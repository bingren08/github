```python
import requests
from bs4 import BeautifulSoup as soup
import time
import pandas as pd
```

```python
headers = {'Accept':'*/*','Accept-Encoding':'gzip, deflate','Accept-Language':'zh-CN,zh;q=0.9','Connection':'keep-alive','User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.79 Safari/537.36'
          }
url = 'http://zdb.pedaily.cn/inv/p2/'
text = requests.get(url,headers=headers).text
print(text)
```

```python
soup_txt = soup(text)
# <ul id="inv-list"> 
inv_list = soup_txt.find('ul',id='inv-list').find_all('li')
print(inv_list[1:][0])
```

```python
a = inv_list[1:][0]
print(a)
```

```python
print(a.find('div',class_='info').text)
print(a.find('dt',class_='company').text)
print(a.find('dt',class_='industry').text)
print(a.find('dt',class_='money').text)

print(a.find('span',class_='r').text)
print(a.find('span',class_='d').text)
print(a.find('span',class_='m').text)


print(a.find('dt',class_='group').text)
```
